```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/home/chengyuli/yanshan/data/DataScience/CCTVNews.csv')
# df = df.dropna()
df.head()
```

```python
df.info()
```

```python
df['time'] = df['time'].apply(lambda x: x.split('：')[-1])
df.head()
```

```python
import re
import urllib.parse

pattern = r"targetpage=(.+?)&"
df['urls-href'] = df['text-href'].apply(lambda x: urllib.parse.unquote(re.search(pattern, x).group(1)))
df.head()
```

```python
import requests
from bs4 import BeautifulSoup as bs
contents = []
for each in df['urls-href']:
    resp = requests.get(each)
    resp.encoding = "utf-8"
    soup = bs(resp.text, "lxml")
    txt = soup.findAll('p')
    content = ''
    for each in txt:
        if each.string:
            content += each.string
    contents.append(content)
df['contents'] = contents
df.head()
```

```python
df = df[['time', 'contents']]
df.head()
```

```python
df.to_csv("/home/chengyuli/yanshan/data/DataScience/rawData/cctv_raw.csv", index=False)
```
